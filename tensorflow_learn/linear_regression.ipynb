```python
import numpy as np

x=np.arange(1,20)
x=x.reshape(19,-1)
```

```python
y=2*x
y=y.reshape(19,-1)
```

```python
bia=np.random.rand(19,1)*5
y=y+bia
```

```python
import matplotlib.pyplot as plot

f=plot.figure()
plot.scatter(x,y)
plot.show()
```

```python
import tensorflow as tf
```

```python
#learn_rate设置0.01时会严重震荡
learn_rate=0.001
max_train_step=10000
```

```python
#构造计算图，y=weight*x+b
X=tf.placeholder(tf.float32,shape=([None,1]),name="x")
#X=tf.Variable(tf.random_normal([1,1]),name="x")
weight=tf.Variable(tf.random_normal([1,1]),name="weight")
b=tf.Variable(tf.zeros([1]),name="bias")
Y=tf.matmul(X,weight)+b
```

```python
Y_=tf.placeholder(tf.float32,shape=([None,1]),name="y")
#损失函数为MSE
loss=tf.reduce_mean(tf.pow(Y-Y_,2))
```

```python
#使用梯度下降优化器
optimizer=tf.train.GradientDescentOptimizer(learn_rate)
train_op=optimizer.minimize(loss)
```

```python
#构造交互式回话
tsess=tf.InteractiveSession()
```

```python
#初始化变量
tsess.run(tf.global_variables_initializer())
# print(weight.eval(),b.eval())
# yr=tsess.run(Y,feed_dict={X:[[1],[2]],Y_:[[2.1],[4.5]]})
# print(yr)
# lossr=tsess.run(loss,feed_dict={X:[[1],[2]],Y_:[[2.1],[4.5]]})
# print(lossr)
# tsess.run(train_op,feed_dict={X:x,Y_:y})
# print(weight.eval(),b.eval())
```

```python
# tsess.run(train_op,feed_dict={X:x,Y_:y})
# print(weight.eval(),b.eval())
# lossr=tsess.run(loss,feed_dict={X:[[1],[2]],Y_:[[2.1],[4.5]]})
# print(lossr)
```

```python
#tsess.run(tf.global_variables_initializer())
#print(X.eval(),weight.eval(),b.eval(),Y.eval())
#训练
for step in range(max_train_step):
    if step % 1==0:
        los=tsess.run(loss,feed_dict={X:x,Y_:y})
        print("step---->{},loss---->{},weight---->{},bias---->{}".format(step,los,weight.eval(),b.eval()))
    tsess.run(train_op,feed_dict={X:x,Y_:y})
```
