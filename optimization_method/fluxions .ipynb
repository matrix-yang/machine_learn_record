```python
import numpy as np

np.set_printoptions(suppress=True)
```

```python
def f(x):
    return x**3+10*x**2+x
```

```python
def f1(x):
    return 3*x**2+20*x+1
```

```python
def f2(x):
    return 6*x+20
```

```python
def f1_tengent(x,xk):
    return f1(xk)+f2(xk)*(x-xk)
```

```python
xs=np.arange(-10,10.01,0.1)
y=np.array([f(x) for x in xs])
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(xs,y)
plt.show()
```

```python
plt.plot(xs,np.array([f1(x) for x in xs]))
plt.show()
```

```python
plt.plot(xs,np.array([f2(x) for x in xs]))
plt.show()
```

```python
s=-1
while(True):
    s=s-f1(s)/f2(s)
    plt.plot(xs,np.array([f(x) for x in xs]))
    plt.plot(xs,np.array([f1(x) for x in xs]))
    plt.plot(xs,np.array([f1_tengent(x,s) for x in xs]))
    plt.plot(xs,np.array([0 for x in xs]))
    plt.scatter(s,f1(s),c='red')
    plt.scatter(s,f1(s))
    plt.show()
    if (abs(f1(s))<0.1):break;
```
