```python
import numpy as np

np.set_printoptions(suppress=True)
```

```python
xs=np.arange(-10,10.01,0.1)
xs,len(xs)
```

```python
y=np.array([x**2 for x in xs])
y,len(y)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
plt.plot(xs,y)
plt.show()
```

```python
def get_gradient(x):
    #学习率
    lambda_=0.8
    return -2*x*lambda_
```

```python
def f(x):
    return x**2
```

```python
def gradient_descent(x):
    y1=f(x)
    delta=get_gradient(x)
    x1=x+delta
    y2=f(x1)
    print("本轮delta:{},x:{}->{},y:{}->{}".format(delta,x,x1,y1,y2))
    return x1,delta
```

```python
x1=10
while(True):
    fig = plt.figure()
    plt.plot(xs,y)
    plt.scatter(x1,f(x1),c='red')
    plt.show()
    x1,delta=gradient_descent(x1)
    if(abs(delta)<0.01):break
```
