```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = "NaN", strategy = "mean", axis = 0)
imputer = imputer.fit(test_df.iloc[ : ,4].values.reshape(-1,1))
test_df.iloc[ : , 4] = imputer.transform(test_df.iloc[ : ,4].values.reshape(-1,1))
```

```python
def discount_gain(row):
    if type(row)==str :
        arr=row.split(':')
        if(len(arr)<2):
            return float(arr[0])
        else:
            weight=round(1-float(arr[1])/float(arr[0]),2)
            return weight
    elif pd.isnull(row) :
        return 1.0

test_df['discount_rate'] = test_df['Discount_rate'].apply(discount_gain)
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = "NaN", strategy = "mean", axis = 0)
imputer = imputer.fit(train_df.iloc[ : ,4].values.reshape(-1,1))
train_df.iloc[ : , 4] = imputer.transform(train_df.iloc[ : ,4].values.reshape(-1,1))
```

```python
train_df['discount_rate'] = train_df['Discount_rate'].apply(discount_gain)
```

```python
train_df.describe()
```

	
消费日期：如果Date=null & Coupon_id != null，该记录表示领取优惠券但没有使用；如果Date!=null & Coupon_id = null，则表示普通消费日期；如果Date!=null & Coupon_id != null，则表示用优惠券消费日期；

```python
def is_use(row):
    Coupon_id = row['Coupon_id']
    Date = row['Date']
    if pd.notnull(Coupon_id)and pd.notnull(Date) :
        return 1
    elif pd.isnull(Coupon_id)and pd.notnull(Date):
        return 2
    elif pd.notnull(Coupon_id) and pd.isnull(Date):
        return 0
```

```python
train_df['use_type'] =train_df.apply(is_use,axis=1)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
fig.set(alpha=0.2)  # 设定图表颜色alpha参数

unuse = train_df.discount_rate[train_df.use_type == 0].value_counts()
use = train_df.discount_rate[train_df.use_type == 1].value_counts()
free = train_df.discount_rate[train_df.use_type == 2].value_counts()
df=pd.DataFrame({'unuse':unuse, 'use':use,'free':free})
df.plot(kind='bar', stacked=True)
plt.title("discount_use")
plt.xlabel("discount_rate") 
plt.ylabel("count") 
plt.show()
```

```python
def calcute(row):
    unuse=row['unuse']
    use=row['use']
    if pd.notnull(unuse) and pd.notnull(use):
        return use/(unuse+use)       
df['rate']=df.apply(calcute,axis=1)
df
```
